```
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
chainingExecutionTime = pd.read_csv("chainingExecutionTimeRecords.csv")
cuckooExecutionTime = pd.read_csv("cuckooExecutionTimeRecords.csv")
```

```
colors = (0,0,0)
area = np.pi*3

plt.scatter(chainingExecutionTime['alpha'], chainingExecutionTime['timeTaken'], s=area, c=colors, alpha=0.5)
plt.title('Hash Chaining Scatter plot for Load Factor with Time Taken(micro secs)')
plt.xlabel('Load Factor (alpha)')
plt.ylabel('Time Taken (micro sec)')
plt.show()
```

```
colors = (0,0,0)
area = np.pi*3

plt.scatter(chainingExecutionTime['alpha'], chainingExecutionTime['accessCount'], s=area, c=colors, alpha=0.5)
plt.title('Hash Chaining Scatter plot for Load Factor with Access Count')
plt.xlabel('Load Factor (alpha)')
plt.ylabel('Access Count')
plt.show()
```

```
colors = (0,0,0)
area = np.pi*3

plt.scatter(cuckooExecutionTime['alpha'], cuckooExecutionTime['timeTaken'], s=area, c=colors, alpha=0.3)
plt.title('Cuckoo Hashing Scatter plot for Load Factor with Time taken (micro sec)')
plt.xlabel('Load Factor (alpha)')
plt.ylabel('Time taken (micro sec)')
plt.show()
```

```
colors = (0,0,0)
area = np.pi*3

plt.scatter(cuckooExecutionTime['alpha'], cuckooExecutionTime['accessCount'], s=area, c=colors, alpha=0.5)
plt.title('Cuckoo Chaining Scatter plot for Load Factor with Access Count')
plt.xlabel('Load Factor (alpha)')
plt.ylabel('Access Count')
plt.show()
```

```
chainAlphas = chainingExecutionTime['alpha']
chainTimeTaken = chainingExecutionTime['timeTaken']
chainAccessCount = chainingExecutionTime['accessCount']

chainAlphas += 1; # to make no alpha value less than 1
```

```
chainAlphas.head()
```

```
plt.figure()
plt.loglog(chainAlphas, chainTimeTaken, ".")

slope, intercept = np.polyfit(np.log10(chainAlphas), np.log10(chainTimeTaken), 1)

x1 = np.log10(np.linspace(1,6, 1000000))
y1 = slope * x1 + intercept
plt.loglog(10**x1, 10**y1)
plt.xlabel("Load Factor")
plt.ylabel("Time Taken (micro second)")
plt.suptitle("Hash Chaining")
```

```
print(slope)
print(intercept)
```

```
plt.figure()
plt.loglog(chainAlphas, chainAccessCount, ".")

slope2, intercept2 = np.polyfit(np.log10(chainAlphas), np.log10(chainAccessCount), 1)

x1 = np.log10(np.linspace(1,6, 1000000))
y1 = slope2 * x1 + intercept2
plt.loglog(10**x1, 10**y1)

plt.xlabel("Load Factor")
plt.ylabel("Total Access")
plt.suptitle("Hash Chaining")
```

```
print(slope2)
print(intercept2)
```

```
cuckooAlphas = cuckooExecutionTime['alpha']
cuckooTimeTaken = cuckooExecutionTime['timeTaken']
cuckooAccessCount = cuckooExecutionTime['accessCount']

cuckooAlphas += 1; # to make no alpha value less than 1
```

```
cuckooAlphas.head()
```

```
plt.figure()
plt.loglog(cuckooAlphas, cuckooTimeTaken, ".")

slope3, intercept3 = np.polyfit(np.log10(cuckooAlphas), np.log10(cuckooTimeTaken), 1)

x1 = np.log10(np.linspace(1,1.5, 1000000))
y1 = slope3 * x1 + intercept3
plt.loglog(10**x1, 10**y1)

plt.xlabel("Load Factor")
plt.ylabel("Time taken (micro seconds)")
plt.suptitle("Cuckoo Hashing")
```

```
print(slope3)
print(intercept3)
```

```
plt.figure()
plt.loglog(cuckooAlphas, cuckooAccessCount, ".")

slope4, intercept4 = np.polyfit(np.log10(cuckooAlphas), np.log10(cuckooAccessCount), 1)

x1 = np.log10(np.linspace(1,1.5, 1000000))
y1 = slope4 * x1 + intercept4
plt.loglog(10**x1, 10**y1)

plt.xlabel("Load Factor")
plt.ylabel("Total Access")
plt.suptitle("Cuckoo Hashing")
```

```
print(slope4)
print(intercept4)
```

```
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

x1 = np.log10(np.linspace(1,6, 1000000))
y1 = slope * x1 + intercept
y2 = slope2 * x1 + intercept2
y3 = slope3 * x1 + intercept3
y4 = slope4 * x1 + intercept4
plt.loglog(10**x1, 10**y1, label="Hash Chaining Alpha vs Time Taken (micros sec)")
plt.loglog(10**x1, 10**y2, label="Hash Chaining Alpha vs Access Count")
plt.loglog(10**x1, 10**y3, label="Cuckoo Hashing Alpha vs Time Taken (micros sec)")
plt.loglog(10**x1, 10**y4, label="Cuckoo Hashing Alpha vs Access Count")
plt.legend()
```
